# Understanding the convergence of central limit theorem

```python
%matplotlib inline

# Import libraries
import numpy as np
from scipy.linalg import eigh
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

***
### Central limit theorem
Let ${x_{1},\ldots, x_{N}}$ be a sample of size $N$, that is a sequence of independent and identically distributed (i.i.d.) random variables drawn from a distribution of the expected value given by $E$ and the variance given by $V$. Then as $N$ approaches infinity, a probability $P(\hat{\theta})$ converges in distribution to a normal:

$$
\begin{equation*}
\lim_{N \to \infty} P(\hat{\theta}) = \sqrt{\frac{N}{2\pi V\left[\phi(x)\right]}}\exp\left(-\frac{N(\hat{\theta} - E\left[\phi(x)\right])^{2}}{2V\left[\phi(x)\right]}\right)
\end{equation*}
$$
where
$$
\begin{equation*}
\hat{\theta} = \frac{1}{N}\Sigma^{N}_{i=1}\phi(x_{i}),
\end{equation*}
$$

$$
\begin{equation*}
E\left[\phi(x)\right] = \int \phi(x)f(x)dx,
\end{equation*}
$$
and
$$
\begin{equation*}
V\left[\phi(x)\right] = E\left[(\phi(x) - E\left[\phi(x)\right])^{2}\right].
\end{equation*}
$$
***

```python
from mpl_toolkits.mplot3d import axes3d

ep = 0 # Expectation
var = 1 # Variance

n = np.arange(5, 100) # Sample size
theta = np.arange(-1.5, 1.5, 0.01) # Average value
N, T = np.meshgrid(n, theta) # Using np.meshgrid()
P = np.sqrt(N / (2 * np.pi * var)) * np.exp(- N * ((T - ep) * (T - ep)) / (2 * var)) # Probability

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(N, T, P, cmap='Spectral_r') # 3D plot
ax.set_xlabel('Sample size')
ax.set_ylabel('Average value')
ax.set_zlabel('Probability')
ax.view_init(17, 200) # Set the point of view
plt.show()
```
