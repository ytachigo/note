## Plot a 2D histgram with log scale

```python
%matplotlib inline
from matplotlib.colors import LogNorm
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(10000)
y = np.random.randn(10000)

plt.hist2d(x, y, bins=100, norm=LogNorm())
plt.colorbar()
```

## Plot multiple graphs at the same time

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 2 * np.pi, 0.1)

plt.figure(figsize=(10, 5))
plt.subplots_adjust(wspace=0.2, hspace=0.5)

for i in range(0, 9):
    plt.subplot(3, 3, i + 1)
    plt.plot(x, np.sin(x))
    plt.plot(x, np.cos(x))
plt.show()
```

## Plot a 3D graph using np.meshgrid()

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

x = np.arange(0, 2 * np.pi, 0.1)
y = np.arange(0, 2 * np.pi, 0.1)
X, Y = np.meshgrid(x, y) # Using np.meshgrid()
Z = np.sin(X) * np.cos(Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='gnuplot') # 3D plot
#ax.plot_wireframe(X, Y, Z) # 3D plot by wireframes
ax.view_init(40, 30) # Set the point of view
plt.show()
```

## Plot a 3D graph using three columns of a DataFrame

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

x = np.arange(0, 2 * np.pi, 0.1)
y = np.arange(0, 2 * np.pi, 0.1)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)
df = pd.DataFrame({'x':X.flatten(), 'y':Y.flatten(), 'z':Z.flatten()})

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(df['x'], df['y'], df['z'], cmap='gnuplot') # 3D plot using three columns of a DataFrame
ax.view_init(40, 30) # Set the point of view
plt.show()
```

## Plot two x- or y- ticks at the same time

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 2 * np.pi, 0.1)
y = np.sin(x)

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.plot(x, y)
ax2 = ax1.twiny() # or twinx()

plt.show()
```

## Control the potision of the legend

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.1)
ya = np.sin(x)
yb = np.cos(x)

plt.plot(x, ya, label='sin')
plt.plot(x, yb, label='cos')
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0, fontsize=18)
```
